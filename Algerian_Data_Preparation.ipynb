```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_update.csv', header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:,"Region"] = 1
```

```python
dataset.info()
```

```python
dataset[["Region"]] = dataset[['Region']].astype(int)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.dropna().reset_index(drop=True)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.iloc[[122]]
```

```python
dataset = dataset.drop(122).reset_index(drop=True)
```

```python
dataset.iloc[[122]]
```

```python
dataset.columns
```

```python
#fix spaces in columnname
dataset.columns = dataset.columns.str.strip()
```

```python
dataset.info()
```

```python
#change the required coloum names
dataset[['month','day','year','Temperature','RH','Ws']] = dataset[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
dataset.info()
```

```python
# change features to float datatype
objects = [features for features in dataset.columns if dataset[features].dtypes == 'O']
```

```python
for i in objects:
  if i!= 'Classes':
    dataset[i]=dataset[i].astype(float)
```

```python
dataset.info()
```

```python
# save the cleaned dataset in csv
dataset.to_csv('cleaned.csv', index=False)
```
